# 1. Pivot()

```python
df_gdp = pd.read_csv('gdp.csv', encoding='latin-1')
df_gdp
```

```python
df_gdp_pivot = df_gdp.pivot(index="year", columns="country", values="gdppc")
df_gdp_pivot
```

## pivot_table()

```python
df_sales = pd.read_excel("supermarket_sales.xlsx")
df_sales
```

```python
df_sales.pivot_table(index='Gender',
                     values=['Quantity', 'Total'],
                     aggfunc='sum').round(1)
```

```python
df_sales.pivot_table(index='Gender',
                     columns='Product line',
                     values='Total',
                     aggfunc='sum').round(1)
```

## Project 2 Veri GÃ¶rselleÅtirme

```python
df_population_raw = pd.read_csv("population_total.csv")
df_population_raw
```

```python
df_population_raw.info()
```

```python
# null olan deÄerleri salÄ±yoruz
df_population_raw.dropna(inplace=True)
```

```python
df_pivot = df_population_raw.pivot(index="year", columns="country",values="population")
df_pivot.head()
```

```python
df_pivot = df_pivot[["United States", "Turkey", "India", "China", "Indonesia"]]
df_pivot
```

## Lineplot

```python
df_pivot.plot(kind="line",
             xlabel="Years", ylabel="Population",
             title="Population (1955-2020)",
             figsize=(8,4))
```

## barplot

```python
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
df_pivot_2020
```

```python
df_pivot_2020 = df_pivot_2020.T
df_pivot_2020
```

```python
df_pivot_2020.plot(kind="bar", color="orange",
                  xlabel="Years", ylabel="Population",
                  title="Population in 2020")
```

## barplot grouped by "n" variables

```python
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
df_pivot_sample.plot(kind="bar",
                  xlabel="Years", ylabel="Population",
                  title="Population in 2020",
                    figsize=(8,4))
```
