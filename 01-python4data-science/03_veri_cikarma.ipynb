# 10. copy()

```python
# reads csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace=True)
# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
df
```

```python
df_copy = df.copy()
df_copy
```

```python
kopyalanmis_veri_m = df_copy.loc["L. Messi", "height_cm"]
orijinal_veri_m = df.loc["L. Messi", "height_cm"]
print(f"Messi boyu\nOrijinal veri: {orijinal_veri} - \tKopyalanmış veri: {kopyalanmis_veri}")
```

```python
df.loc["L. Messi", "height_cm"] = 180

print(f"Messi boyu\nOrijinal veri: {orijinal_veri_m} - \tKopyalanmış veri: {kopyalanmis_veri_m}")
```

```python
df_shallow_copy = df.copy(deep=False)
kopyalanmis_veri = df_shallow_copy.loc["Cristiano Ronaldo", "height_cm"]
orijinal_veri = df.loc["Cristiano Ronaldo", "height_cm"]
print(f"Ronaldo boyu\nOrijinal veri: {orijinal_veri} - \tKopyalanmış veri: {kopyalanmis_veri}")
```

```python
df.loc["Cristiano Ronaldo", "height_cm"] = 200
```

```python
kopyalanmis_veri = df_shallow_copy.loc["Cristiano Ronaldo", "height_cm"]
orijinal_veri = df.loc["Cristiano Ronaldo", "height_cm"]
print(f"Ronaldo boyu\nOrijinal veri: {orijinal_veri} - \tKopyalanmış veri: {kopyalanmis_veri}")
```
