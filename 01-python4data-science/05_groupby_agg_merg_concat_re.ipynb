# Birleştirme (concat)

```python
df_movies = pd.read_csv('IMDb movies.csv', low_memory=False)
df_ratings = pd.read_csv('IMDb ratings.csv')
```

```python
# select columns
df_movies = df_movies[['imdb_title_id', 'title', 'year',
                       'genre', 'country', 'director', 'actors']]

df_ratings = df_ratings[['imdb_title_id', 'total_votes', 'mean_vote']]
```

## Concatenate vertically

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['E', 'F', 'G', 'F'],
                    'age': [40, 21, 19, 24]})

pd.concat([df1, df2])
```

## Concatenate horizontally

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'job': ['Doctor', 'Statistician',
                            'Accountant', 'Developer']})

pd.concat([df1, df2], axis=1)
```

```python
# shape of dataframes that we'll concatenate
print(df_movies.shape)
print(df_ratings.shape)
```

```python
# concatenate df_movies and df_ratings on 'imdb_title_id' (horizontally along the columns)
df_concat_horizontally = pd.concat([
    df_movies.set_index('imdb_title_id'), df_ratings.set_index('imdb_title_id')
], axis=1)

print(df_concat_horizontally.shape)
```

# Joins

## merge() : ortak sütun veya indeks

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['C', 'D', 'E', 'F'],
                    'job': ['Doctor', 'Statistician',
                            'Accountant', 'Developer']})
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, on='id', how='inner')
```

```python
df1.merge(df2, on='id', how='outer')
```

```python
df1.merge(df2, on='id', how='outer')
```

```python
df1.merge(df2, on='id', how='outer',
          indicator=True)
```

```python
df1.merge(df2, on='id', how='outer',
          indicator=True)
```

# Regular expressions

```python
import re
```

## re.search()

```python
text = """Hello World.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""
```

```python
re.search(r'\d', text) # digit
```

```python
re.search(r'^hello', text, flags=re.M)
```

## re.findall()

```python
re.findall(r'\d', text) # tüm sayıları buluyor
```

## Egzersiz

```python
# Match right date format (MM-DD-YYYY)
text = '''
13-04-2021
2021-13-04
2021-04-13
'''
re.findall(r'\d{2}-\d{2}-\d{4}', text)
```

```python
text = """Hello World.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""
re.findall(r'[^\w\s]', text)
```
